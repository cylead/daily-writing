```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import multivariate_normal
import seaborn as sns
```

### Sample from the multidimentional normal distribution

```python
cov = np.array([[6, -3], [-3, 3.5]])
pts = np.random.multivariate_normal([0, 0], cov, size=5000)
pts.mean(axis=0), np.cov(pts.T)
```

#### Scatter plot

```python
plt.plot(pts[:, 0],pts[:, 1], 'x')
plt.axis('equal')
plt.show()
```

```python
#### marginal distribution esimation
```

```python
sns.displot(pts[:, 0], bins=10, kde=True)
plt.show()
```

### Demonstrate the linear combination of each dimentional is still normal distributed

```python
dim = 5
# random mean vector
mean = np.random.rand(dim)
# random psd cov matrix
rand = np.random.rand(dim, dim)
cov = rand @ rand.T
mnd = np.random.multivariate_normal(mean, cov, size=2000)
# mnd.mean(axis=0) - mean, np.cov(mnd.T) - cov
```

```python
# random linear combination coefficients
A = np.random.rand(dim)
linear_comb = mnd @ A
sns.displot(linear_comb, bins=10, kde=True)
plt.show()
```

### The sum of two normal distributed RV is still normal distribued

```python
nd1 = np.random.normal(1,2, size=500)
nd2 = np.random.normal(3,5, size=500)
coeff = np.random.rand(2)
nd_sum = coeff[0] * nd1 + coeff[1] * nd2
sns.displot(nd_sum, bins=10, kde=True)
plt.show()
```
