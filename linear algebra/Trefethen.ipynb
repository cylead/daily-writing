```python
a = np.array([[0, 1], [3, -1]])
u, s, vh = np.linalg.svd(a, full_matrices=True)

u1 = u[:,0]
u2 = u[:,1]
svh = np.diag(s)@vh
v1 = svh[0, :]
v2 = svh[1, :]                 # in degrees
angle_v = np.angle(v1[0]+v1[1]*1j, deg=True)    
```

```python
plt.figure()
ax = plt.gca()

ellipse_u = Ellipse(xy=(0, 0), width=2, height=2, edgecolor='k', fc='None')
ax.add_patch(ellipse_u)

ellipse_v = Ellipse(xy=(0, 0), width=2*s[0], height=2*s[1], edgecolor='k', fc='None', angle = angle_v)
ax.add_patch(ellipse_v)

plt.arrow(0, 0, u1[0], u1[1], width = 0.02, length_includes_head=True)
plt.arrow(0, 0, u2[0], u2[1], width = 0.02, length_includes_head=True)

plt.arrow(0, 0, v1[0], v1[1], width = 0.02, length_includes_head=True)
plt.arrow(0, 0, v2[0], v2[1], width = 0.02, length_includes_head=True)
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.grid()
plt.gca().set_aspect("equal")
plt.show()
```
